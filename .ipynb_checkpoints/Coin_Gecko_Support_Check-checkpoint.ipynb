```python
# Method 2 (better, but could fail, manual check needed!)
#
# dynamic scraps the html from etoro, returns parsed coin symbol list and coin name list
# 
# Pre-condition: chrome and its corresponding chromedriver.exe are installed, 
#                correct chromedriver.exe path is provided in steps 2, have good internet connection 
# Post-condition: correct coin symbols and names are parsed

def etoro_web_pages_to_lists():
    # 1. Configures webdriver
    #
    # Hides GUI can solve the problem of web page stucking on loadding

    options = Options()
    options.headless = True # Hides GUI

    # Sets window size to native GUI size
    # options.add_argument("--window-size=1920,1080") 
    # options.add_argument("start-maximized")

    # Removes automation tags from the browser
    # options.add_experimental_option("excludeSwitches", ["enable-automation"])
    # options.add_experimental_option('useAutomationExtension', False)

    # Reference:
    # https://scrapfly.io/blog/web-scraping-with-selenium-and-python/
    # https://python.plainenglish.io/how-to-bypass-google-recaptcha-with-selenium-and-python-9470664fb575
    
    
    # 2. Initializes driver

    service = Service('C:/Users/ZH128895/chromedriver.exe') # Your/Absolute/Path/To/chromedriver.exe 
    driver = webdriver.Chrome(service=service, options = options) # Uses Chrome brower

    # Changes User Agent at each request, cannot execute without this line, not sure why 
    driver.execute_cdp_cmd('Network.setUserAgentOverride', {"userAgent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.53 Safari/537.36'})

    driver.get('https://www.etoro.com/discover/markets/cryptocurrencies/coins') # Website to scrap
    
    
    # Finds two buttons elements by XPATH

    next_page_btn = driver.find_element(By.XPATH, '/html/body/ui-layout/div/et-layout-main/div/div[2]/div[2]/div[3]/div/ng-view/et-discovery-markets-results/div/et-discovery-markets-results-header/div/div[2]/div/div[2]/a[2]')

    prev_page_btn = driver.find_element(By.XPATH, '/html/body/ui-layout/div/et-layout-main/div/div[2]/div[2]/div[3]/div/ng-view/et-discovery-markets-results/div/et-discovery-markets-results-header/div/div[2]/div/div[2]/a[1]')
    
    
    # Navigates to next page then return back
    #
    # The reason to perform this action is there is one coin item not loaded the first time loads to the site,
    # the missing coin item is loaded after navigating to the next page then navigates back

    # Creates action chain object
    action = ActionChains(driver)

    # Clicks the next page button
    action.click(on_element = next_page_btn)

    # Clicks the previous page button
    action.click(on_element = prev_page_btn)

    # Performs the operation
    action.perform()
    
    
    # Extracts the coin info text


    # Init two lists for storing coin symbol and coin name respectively
    coin_symbol_lst = []
    coin_name_lst = []

    # Scraps the coin info from the first page
    coin_symbols_1 = driver.find_elements(By.CSS_SELECTOR, 'div.symbol')
    coin_names_1 = driver.find_elements(By.CSS_SELECTOR, 'div.name')

    for index in range(len(coin_symbols_1)):
        coin_symbol_lst.append(coin_symbols_1[index].text)
        coin_name_lst.append(coin_names_1[index].text)

    # Navigates to next page
    action = ActionChains(driver)

    action.click(on_element = next_page_btn)

    action.perform()

    # Scraps the coin info from the second page
    coin_symbols_2 = driver.find_elements(By.CSS_SELECTOR, 'div.symbol')
    coin_names_2 = driver.find_elements(By.CSS_SELECTOR, 'div.name')

    for index in range(len(coin_symbols_2)):
        coin_symbol_lst.append(coin_symbols_2[index].text)
        coin_name_lst.append(coin_names_2[index].text)
        
    return coin_symbol_lst, coin_name_lst
```

# Perfoems Gecko Support Check in Etoro

```python
# Reads excel, converts to dataframe

df_coin_etoro = pd.read_excel('CoinEtoro.xlsx')

df_coin_gecko = pd.read_excel('CoinGecko.xlsx')
```

```python
# Extracts cols in Etoro, converts them to list

df_etoro_symbol = df_coin_etoro.loc[:, 'Coin(Etoro) Symbol']
df_etoro_name = df_coin_etoro.loc[:, 'Coin(Etoro) Name']
etoro_symbol_lst = df_etoro_symbol.tolist()
etoro_name_lst = df_etoro_name.tolist()
```

```python
# Extracts cols in Gecko, converts them to list

df_gecko_symbol = df_coin_gecko.loc[:, 'symbol']
df_gecko_name = df_coin_gecko.loc[:, 'name']
gecko_symbol_lst = df_gecko_symbol.tolist()
gecko_name_lst = df_gecko_name.tolist()
```

```python
# Performs the coin gecko support checks
#
# If etoro symbol is exactly equivalent to gecko symbol, stores only(or all) the exact matches
# Otherwise (no exact match), if etoro symbol is a substring of gecko symbol, stores all the substring matches

has_gecko_support = []
gecko_ref_map = []

# Iterates through [etoro_symbol_lst]
for i in range(len(etoro_symbol_lst)):
    
    # Extracts this current etoro symbol to be checked
    etoro_symbol = etoro_symbol_lst[i]
    
    # Initializes them as No/None
    has_gecko_support.append('No')
    gecko_ref_map.append({})
    
    # Temp var to store has exact match
    has_exact_match = False
    
    # Iterates through [gecko_symbol_lst], looks for exact matches 
    for j in range(len(gecko_symbol_lst)):
        
        # Extacrs this current gecko symbol to check for support
        gecko_symbol = gecko_symbol_lst[j]
        
        # Has exact match
        if str(etoro_symbol).lower() == str(gecko_symbol).lower():
            has_gecko_support[i] = ('Yes')
            
            # Checks if there's multiple same keys
            if gecko_symbol_lst[j].upper() in gecko_ref_map[i]:
                # If so, appends the gecko name value
                gecko_ref_map[i][gecko_symbol_lst[j].upper()].append(gecko_name_lst[j])
            else:
                # If not, init a new gecko symbol-name pair as a key-value pair
                gecko_ref_map[i][gecko_symbol_lst[j].upper()] = [gecko_name_lst[j]]
            
            has_exact_match = True
            
    # Iterate through [gecko_symbol_lst] only if there is no exact match, mark the [has_gecko_support] as 'No'
    if (not has_exact_match):
        for k in range(len(gecko_symbol_lst)):
            
            # Extracts current gecko symbol to check for support
            gecko_symbol = gecko_symbol_lst[k]
        
            if str(etoro_symbol).lower() in str(gecko_symbol).lower():
                has_gecko_support[i] = ('No')
                
                # Checks if there's multiple same keys
                if gecko_symbol_lst[k].upper() in gecko_ref_map[i]:
                    # If so, appends the gecko name value
                    gecko_ref_map[i][gecko_symbol_lst[k].upper()].append(gecko_name_lst[k])
                else:
                    # If not, init a new gecko symbol-name pair as a key-value pair
                    gecko_ref_map[i][gecko_symbol_lst[k].upper()] = [gecko_name_lst[k]]
```

```python
# Creates dataframe, export as excel

result_data = {
    'Coin(Etoro) Symbol' : etoro_symbol_lst,
    'Coin(Etoro) Name' : etoro_name_lst,
    'Has Gecko Support' : has_gecko_support,
    'Gecko Coin Reference (Symbol : [Name])' : gecko_ref_map
}

df_results = pd.DataFrame(result_data)

df_results.to_csv('C:/Users/ZH128895/Work/CoinGeckoSupportCheck/CoinGeckoSupportCheckResults/CoinGeckoSupportCheck_6-21-456.csv', index = False)
```
